## importing Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas
import re
import time
```

## copying url from the flipkart.com


```python
url="https://www.flipkart.com/food-products/pr?sid=eat&fm=neo%2Fmerchandising&iid=M_776fdd1a-3b72-44cd-bc4e-4df6d71d496a_1_X1NCR146KC29_MC.Y08EI5EDJF01&otracker=hp_rich_navigation_3_1.navigationCard.RICH_NAVIGATION_Beauty%2C%2BFood..~Food%2B%26%2BDrinks_Y08EI5EDJF01&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_3_L1_view-all&cid=Y08EI5EDJF01"
```

## Getting response/requests/status code

```python
response=requests.get(url)
response.status_code
```

```python
request_header = {'Content-Type': 'image/webp; charset=UTF-8', 
                  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:126.0) Gecko/20100101 Firefox/126.0',
                'Accept-Encoding': 'gzip, deflate, br, zstd'                  
                 }

response = requests.get(url, headers=request_header)
```

## getting html Text

```python
soup=BeautifulSoup(response.text)
soup
```

```python
soup.text
```

## products_Names

```python
product_name=[]
name=soup.find_all('a',class_='wjcEIp')
for i in name:
    product_name.append(i.text)
print(product_name)    
```

## product_price

```python
product_prices=[]
prices=soup.find_all('div',class_='Nx9bqj')
for i in prices:
    product_prices.append(i.text)
print(product_prices)    
```

## product_weights

```python
product_weights=[]
weights=soup.find_all('div',class_='NqpwHC')
for i in weights:
    product_weights.append(i.text)
print(product_weights)   
```

## product_reviews

```python
product_reviews=[]
reviews=soup.find_all('div',class_='XQDdHH')
for i in reviews:
    product_reviews.append(i.text)
print(product_reviews)    
```

## product_Discount

```python
product_Discount=[]
Discount=soup.find_all('div',class_="UkUFwK")
for i in Discount:
    product_Discount.append(i.text)
print(product_Discount)    
```

## product_offers

```python
product_offers=[]
offers=soup.find_all('div',class_="yiggsN O5Fpg8")
for i in offers:
    product_offers.append(i.text)
print(product_offers)    
```

## products Length

```python
print("Product Name:\n",len(product_name))
print("Product Price:\n",len(product_prices))
print("Product weigths:\n",len(product_weights))
print("product reviews:\n",len(product_reviews))
print("product Discount:\n",len(product_Discount))
print("product offers:\n",len(product_offers))
      
```

```python
min_len = min(
    len(product_name),
    len(product_prices),
    len(product_weights),
    len(product_reviews),
    len(product_Discount),
    len(product_offers))

product_name = product_name[:min_len]
product_prices = product_prices[:min_len]
product_weights = product_weights[:min_len]
product_reviews = product_reviews[:min_len]
product_Discount = product_Discount[:min_len]
product_offers = product_offers[:min_len]

import pandas as pd

df = pd.DataFrame({
    "Product Name": product_name,
    "Product Prices": product_prices,
    "Product Weights": product_weights,
    "Product Reviews": product_reviews,
    "Product Discount": product_Discount,
    "Product Offers": product_offers
})

df
```

```python
df.to_csv("scraped_products.csv",index=False)
print("Data saved as scraped_products.csv")
```

## using  regex removing product Name after dash

```python

df['Product Name'] = df['Product Name'].str.replace(r'\s*[-].*', '', regex=True).str.strip()
df
```

```python
df['Product Prices'] = (df
    ['Product Prices']
    .str.replace('', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
df
```

```python
df['Product Discount'] = df['Product Discount'].str.replace(r'%\s*off', '', regex=True).str.strip()
df
```

## using database making data visulaitions by bivariant

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read your dataset
df = pd.read_csv("scraped_products.csv")   # or your dataframe variable

# Clean columns
df.columns = df.columns.str.strip()

```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
plt.figure(figsize=(20,20))
sns.scatterplot(x='Product Prices', y='Product Discount', data=df, hue='Product Reviews', palette='cool')
plt.title("Price vs Discount (colored by Rating)")
plt.xlabel("Product Price")
plt.ylabel("Product Discount")

plt.gca().set_xticklabels([f"{int(x)}" for x in plt.gca().get_xticks()])
plt.gca().set_yticklabels([f"{int(y)}%" for y in plt.gca().get_yticks()])

plt.show()

```

```python
plt.figure(figsize=(9,9))
sns.boxplot(x='Product Reviews', y='Product Discount', data=df, palette='Set2')
plt.title("Discount offered across Ratings")
plt.xlabel("Review Rating")
plt.ylabel("Discount (%)")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='Product Reviews', y='Product Prices', data=df, palette='coolwarm')
plt.title("Price variation across Ratings")
plt.xlabel("Review Rating")
plt.ylabel("Product Price ()")
plt.show()

```

## univarite

```python
plt.figure(figsize=(10,10))
plt.pie(df['Product Prices'].value_counts(),
        labels=df['Product Prices'].value_counts().index,
        autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Product Prices ()")
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Product Discount'], bins=10, color='orange', kde=True)
plt.title("Distribution of Product Discounts (%)")
plt.xlabel("Discount (%)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Product Reviews', data=df, palette='mako')
plt.title("Product Review Ratings Count")
plt.xlabel("Ratings")
plt.ylabel("Number of Products")
plt.show() 
```
